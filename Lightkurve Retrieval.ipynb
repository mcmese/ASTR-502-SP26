```python
import lightkurve as lk
import numpy as np
import pandas as pd
import matplotlib as pyplot

#To search for one target individually
#search_result = lk.search_lightcurve('TIC 201248411', mission ='TESS')
#search_result

#Read in csv file
df = pd.read_csv('ASTR502_Mega_Target_List.csv')

# Loop through each target
for target in df["tic_id"]:
    print(f"\nProcessing {target}")

    try:
        search_result = lk.search_lightcurve(target, mission="TESS", author="SPOC")

        if len(search_result) == 0:
            print("  No light curves found.")
            continue

        # Download all available sectors
        lc_collection = search_result.download_all()

        print(f"  Downloaded {len(lc_collection)} light curves")

    except Exception as e:
        print(f"  Error processing {target}: {e}")
```
